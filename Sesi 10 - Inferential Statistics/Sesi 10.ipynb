## Uniform Distribution

```python
from scipy.stats import uniform
```

```python
n = 10000
start = 10
width = 20
data_uniform = uniform.rvs(size=n, loc = start, scale=width)
```

```python
ax = sns.distplot(data_uniform,
                  bins=100,
                  kde=True,
                  color='skyblue',
                #   hist_kws={"linewidth": 15,'alpha':1})
)
ax.set(xlabel='Uniform Distribution ', ylabel='Frequency')
```

# CI for the population Proportion in Python

```python
import pandas as pd
import numpy as np

df = pd.read_csv('D:\STATISTIKA\BISMILLAH\KM\SI\Hacktiv8\Sesi 10 - Inferential Statistics\Heart.csv')
df
```

```python
df['Sex1'] = df.sex.replace({1: "Male", 0: "Female"})
df['Sex1'].unique()
```

```python
df.head()
```

```python
dx = df[["target", "Sex1"]].dropna()
dx.head()
```

```python
pd.crosstab(dx.target, dx.Sex1)
```

```python
p_fm = 226/(86+226)
p_fm
```

```python
n = 86+226
n
```

```python
# standar error
se_female = np.sqrt(p_fm * (1 - p_fm) / n)
se_female
```

```python
# CI
z_score = 1.96

lcb = p_fm - z_score* se_female #lower 
ucb = p_fm + z_score* se_female #upper 

lcb, ucb
```

```python
import statsmodels.api as sm
sm.stats.proportion_confint(n * p_fm, n)
```

## Difference in Population Means

```python
url = "https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/nhanes_2015_2016.csv"
da = pd.read_csv(url)

da.head()
```

```python
females = da[da["RIAGENDR"] == 2]
male = da[da["RIAGENDR"] == 1]
n1 = len(females)
mu1 = females["BMXBMI"].mean()
sd1 = females["BMXBMI"].std()

(n1, mu1, sd1)
```

```python
n2 = len(male)
mu2 = male["BMXBMI"].mean()
sd2 = male["BMXBMI"].std()

(n2, mu2, sd2)
```

```python
sm.stats.ztest(females["BMXBMI"].dropna(), 
male["BMXBMI"].dropna(),alternative='two-sided')
```

### Conclusion of the hypothesis test

```python
plt.figure(figsize=(7,4))
plt.title("Female BMI histogram",fontsize=16)
plt.hist(females["BMXBMI"].dropna(),edgecolor='k',color='pink',bins=25)
plt.show()

plt.figure(figsize=(7,4))
plt.title("Male BMI histogram",fontsize=16)
plt.hist(male["BMXBMI"].dropna(),edgecolor='k',color='blue',bins=25)
plt.show()
```

# Hypothesis Testing for One Proportion

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import scipy.stats.distributions as dist
df = pd.read_csv('D:\STATISTIKA\BISMILLAH\KM\SI\Hacktiv8\Sesi 10 - Inferential Statistics\Heart.csv')

df.head()
```

```python
p_us = len(df[df['target']=='1'])/len(df)
p_us
```

```python
se = np.sqrt(0.42 * (1-0.42) / len(df))
se
```

```python
be = p_us

he = 0.42
test_stat = (be - he)/se
test_stat
```

```python
pvalue = 2*dist.norm.cdf(-np.abs(test_stat))
pvalue
```
